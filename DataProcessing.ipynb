```python
import os
from google.colab import drive
```

```python
import librosa 
import librosa.display
import math
import json
```

```python
drive.mount('/content/drive')
```

```python
dataset_path = "/content/drive/MyDrive/genres"
json_path = "/content/data.json"

SAMPLE_RATE = 22050
DURATION = 30
samples_per_track = SAMPLE_RATE*DURATION
```

```python
def save_mfcc(datset_path,json_path,n_mfcc=13,n_fft=2048,hop_length=512,num_segments=5):

  #Storing Data in form of Dictionary
  data ={
      "mapping":[],
      "mfcc":[],
      "labels":[]
  }

  
  num_samples_per_segment = int(samples_per_track/num_segments)
  expected_num_vectors = math.ceil(num_samples_per_segment/hop_length)

  for i,(dirpath,dirnames,filenames) in enumerate(os.walk(dataset_path)):
    if dirpath is not dataset_path:


      dirpath_components = dirpath.split('/')
      semantic_label = dirpath_components[-1]
      data["mapping"].append(semantic_label)
      print("Processing:{}".format(semantic_label))

      for file in filenames:
        file_path = os.path.join(dirpath,file)
        audio,sr = librosa.load(file_path,sr=SAMPLE_RATE)

        for sample in range(num_segments):
          start_sample = num_samples_per_segment*sample
          finish_sample = start_sample + num_samples_per_segment

          mfcc = librosa.feature.mfcc(audio[start_sample:finish_sample],sr = sr,n_fft=n_fft,n_mfcc=n_mfcc,hop_length=hop_length)
          mfcc = mfcc.T
          
          if len(mfcc)==expected_num_vectors:
            data["mfcc"].append(mfcc.tolist())
            data["labels"].append(i-1)
            print("{},segment:{}".format(file_path,sample+1))

  with open(json_path,"w") as fp:
    json.dump(data,fp,indent=4)

if __name__ == '__main__':
  save_mfcc(dataset_path,json_path,num_segments=10)



```
